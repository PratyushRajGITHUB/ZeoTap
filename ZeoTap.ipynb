```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import davies_bouldin_score, silhouette_score
```

```python
dummy_features = customer[['CustomerID']]
```

```python
dummy_features = pd.get_dummies(customer[['CustomerID', 'Region']], columns=['Region'])
```

```python
similarity_matrix = cosine_similarity(dummy_features.drop(columns=['CustomerID']).values)
similarity_df = pd.DataFrame(similarity_matrix, index=customer['CustomerID'], columns=customer['CustomerID'])
```

```python
lookalike_results = {}
for customer_id in customer['CustomerID'][:20]:
    similar_customer = similarity_df[customer_id].sort_values(ascending=False).iloc[1:4]
    lookalike_results[customer_id] = list(similar_customer.items())
```

```python
lookalike_df = pd.DataFrame.from_dict(lookalike_results, orient='index')
lookalike_df.columns = ['Lookalike_1', 'Lookalike_2', 'Lookalike_3']
lookalike_df.to_csv('FirstName_LastName_Lookalike.csv', index_label='CustomerID')
```

```python
customer_data = pd.merge(customer, transactions, on='CustomerID', how='inner')
```

```python
customer_summary = customer_data.groupby('CustomerID').agg({
    'TotalValue': 'sum',  # Total spending by the customer
    'TransactionID': 'count',  # Number of transactions
    'Quantity': 'sum',  # Total quantity purchased
}).rename(columns={
    'TotalValue': 'TotalSpending',
    'TransactionID': 'TransactionCount',
    'Quantity': 'TotalQuantity'
}).reset_index()
```

```python
customer_profile = pd.merge(customer, customer_summary, on='CustomerID', how='inner')
```

```python
customer_profile = pd.get_dummies(customer_profile, columns=['Region'], drop_first=True)
```

```python
scaler = StandardScaler()
numerical_features = ['TotalSpending', 'TransactionCount', 'TotalQuantity']
customer_profile[numerical_features] = scaler.fit_transform(customer_profile[numerical_features])
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score, silhouette_score

# Define numerical features
numerical_features = ['TotalSpending', 'TransactionCount', 'TotalQuantity']

# Standardize the numerical features
scaler = StandardScaler()
customer_profile[numerical_features] = scaler.fit_transform(customer_profile[numerical_features])

# Initialize lists to store evaluation metrics
db_indexes = []
silhouette_scores = []

# Perform clustering for k values from 2 to 10
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(customer_profile[numerical_features])  # Fit and predict clusters
    db_index = davies_bouldin_score(customer_profile[numerical_features], labels)  # DB Index
    silhouette = silhouette_score(customer_profile[numerical_features], labels)  # Silhouette Score
    db_indexes.append(db_index)
    silhouette_scores.append(silhouette)

# Print the results
print("Davies-Bouldin Indexes:", db_indexes)
print("Silhouette Scores:", silhouette_scores)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), db_indexes, marker='o')
plt.title('Davies-Bouldin Index for Different Cluster Sizes')
plt.xlabel('Number of Clusters')
plt.ylabel('Davies-Bouldin Index')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', color='green')
plt.title('Silhouette Score for Different Cluster Sizes')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```
